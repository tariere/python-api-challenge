#### Objective
Your objective is to build a series of scatter plots to showcase the following relationships:

* Temperature (F) vs. Latitude
* Humidity (%) vs. Latitude
* Cloudiness (%) vs. Latitude
* Wind Speed (mph) vs. Latitude

Your final notebook must:

* Randomly select **at least** 500 unique (non-repeat) cities based on latitude and longitude.
* Perform a weather check on each of the cities using a series of successive API calls.
* Include a print log of each city as it's being processed with the city number and city name.
* Save both a CSV of all data retrieved and png images for each scatter plot.

```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from datetime import date
import os

# Import API key
from api_keys import api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output Files (CSV)
weather_data_files = "output_data/cities.csv"
# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
```

```python
#take a peek at my cities
cities
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
#store the base url in a variable so that it can be used in my for loop
base_url = "http://api.openweathermap.org/data/2.5/weather?units=Imperial&q="

#create a list that will hold my dictionaries
cities_data = []

#set up some counters for the record and set printouts
rec_count = 0
set_count = 0

    
print("Beginning Data Retrieval")  
print("------------------------")
#for loop to do the follow steps
for city in (cities):
    #increase the counters based on the following conditions
    if rec_count == 0 or rec_count < 50 :
        rec_count +=1
    elif rec_count == 50:
        set_count +=1
        rec_count = 1
        
    #update the URL for each city in the list
    url = base_url + city.replace(" ", "+") + "&appid="+api_key
    
    #print out the log as each city is being processed
    print(f"Processing Record {rec_count} of Set {set_count} |{city}")
    
    #process the URL and convert to JSON, then place in a variable
    post_response = requests.get(url).json()
    
    #print the URL for the log
    print(url)
    
    #in a try and except/loop, extract elements of the JSON object and place those objects in a dictionary, and each dictionary will be added to a list. 
    try:
        cities_data.append({
            "City" : city.capitalize(),
            "Cloudiness" : post_response["clouds"]["all"],
            "Country" : post_response["sys"]["country"],
            "Date" : post_response["dt"],
            "Humidity" : post_response["main"]["humidity"],
            "Lat" : post_response["coord"]["lat"],
            "Lng" : post_response["coord"]["lon"],
            "Max Temp" : post_response["main"]["temp_max"],  
            "Wind Speed" : post_response["wind"]["speed"]
          })
        
    except:
        print("City not found. Skipping...")
        pass
print("-----------------------")
print("Data Retrieval Complete")
print("-----------------------")

```

```python
#take a peek a my list of dictionaries
cities_data
```

### Convert Raw Data to DataFrame
* Export the city data into a .csv.
* Display the DataFrame

```python
#take my list of dicts and convert to a dataframe
cities_df = pd.DataFrame(cities_data)
cities_df.head()
```

```python
#export the dataframe to a csv
cities_df.to_csv(weather_data_files)
```

```python
#quick count of records in my dataframe
cities_df.count()
```

### Plotting the Data
* Use proper labeling of the plots using plot titles (including date of analysis) and axes labels.
* Save the plotted figures as .pngs.

#### Latitude vs. Temperature Plot

```python
#capture the date the report was run
today = date.today()
```

```python
#put the latitudes and the temps in variables for plotting
c_lats = cities_df["Lat"]
c_temps = cities_df["Max Temp"]

#plot the data
fig, ax = plt.subplots()
ax.scatter(c_lats,c_temps,edgecolors="k", c = "b");
ax.set_xlabel('Latitude');
ax.set_ylabel('Max Tempurature (F)');
ax.set_title(f'City Latitude vs Max Tempurature {today.strftime("%m/%d/%Y")}');
ax.grid(color='w');
ax.set_facecolor('#DFDFDF');
ax.set_ylim([-100,150]);
ax.set_xlim([-80,100]);
             
fig.savefig(os.path.join("plots","01_lat_v_temp.png"))
```

#### Latitude vs. Humidity Plot

```python
#put the humidity data in a variable for plotting
c_hum = cities_df["Humidity"]

#plot the data
fig, ax = plt.subplots()
ax.scatter(c_lats,c_hum,edgecolors="k", c = "b");
ax.set_xlabel('Latitude');
ax.set_ylabel('Humidity%');
ax.set_title(f'City Latitude vs Humidity {today.strftime("%m/%d/%Y")}');
ax.grid(color='w');
ax.set_facecolor('#DFDFDF');
ax.set_ylim([-20,120]);
ax.set_xlim([-80,100]);
             
fig.savefig(os.path.join("plots","02_lat_v_hum.png"))
```

#### Latitude vs. Cloudiness Plot

```python
#put the Coudiness data in a variable for plotting
c_cloud = cities_df["Cloudiness"]

#plot the data
fig, ax = plt.subplots()
ax.scatter(c_lats,c_cloud,edgecolors="k", c = "b");
ax.set_xlabel('Latitude');
ax.set_ylabel('Cloudiness%');
ax.set_title(f'City Latitude vs Cloudiness {today.strftime("%m/%d/%Y")}');
ax.grid(color='w');
ax.set_facecolor('#DFDFDF');
ax.set_ylim([-20,120]);
ax.set_xlim([-80,100]);
             
fig.savefig(os.path.join("plots","03_lat_v_cloud.png"))
```

#### Latitude vs. Wind Speed Plot

```python
#put the windspeeed data in a variable for plotting
c_wind = cities_df["Wind Speed"]

#plot the data
fig, ax = plt.subplots()
ax.scatter(c_lats,c_wind,edgecolors="k", c = "b");
ax.set_xlabel('Latitude');
ax.set_ylabel('Wind Speed (mph');
ax.set_title(f'City Latitude vs Windspeed {today.strftime("%m/%d/%Y")}');
ax.grid(color='w');
ax.set_facecolor('#DFDFDF');
ax.set_ylim([-5,40]);
ax.set_xlim([-80,100]);
             
fig.savefig(os.path.join("plots","04_lat_v_wind.png"))
```

# Conclusions

In looking at the plots of the data that was obtained, I can make the following observations

**1. The closer a location is to the equator, the warmer the temperature.**
There is a trail of data points from -60 latitude where the temperatures are rising.  When they get to about -20 to 20 latitude, the temperatures peak before starting to drop back down again after 20 latitude.  Since the equator is at 0, it makes sense the to say that the the temperatures get warmer as you get closer to the equator. As you get away from the equator, the temperatures tend to come back down. 

**2. If there's measurable humidity, it appears to be clustered around the 60%-100% range**
It seems that if areas have humidity, then they have a lot of humidity.  There are some areas that have lower humidity, but the concentration of data with humidity values are all between 60%-100%. This does not appear to be impacted by the latitude, as these values span across the latitude range. 

**3. There does not seem to be much correlation between latitude and cloudiness or wind speed**
This is based on the fact that the data does not appear to be clustered in any particular area/latitude for these dimensions (wind speed or cloudiness).  For Cloudiness, there are some high data points across the latitude axis, meaning that places all over the world experience cloudiness.  For Wind speed, the data points are concentrated around speeds of 0-15 mph, but again, it's spread across the latitude axis. There are some high wind speed data points, but those are likely outliers, as opposed to an observable trend.
